```python
cols = ['inv_01_dc_current_inv_149579'	,'inv_01_dc_voltage_inv_149580',	'inv_01_ac_current_inv_149581',	'inv_01_ac_voltage_inv_149582', 'inv_01_ac_power_inv_149583']
# Convert 'measured_on' to datetime and set it as index
elec_filtered_new = elec_filtered[cols].copy()  # Create a copy to avoid SettingWithCopyWarning
elec_filtered_new['measured_on'] = pd.to_datetime(elec_filtered['measured_on'])
elec_filtered_new.set_index('measured_on', inplace=True)
elec_filtered_new.head()
```

```python
import pandas as pd

# Perform sequential merges on 'measured_on'
# Start by merging elec_filtered and env_filtered
merged_df = pd.merge(elec_filtered, env_filtered, on='measured_on', how='inner')

# Merge the result with meter_filtered
merged_df = pd.merge(merged_df, meter_filtered, on='measured_on', how='inner')

# Finally, merge the result with irr_filtered
merged_df = pd.merge(merged_df, irr_filtered, on='measured_on', how='inner')

# Display the final merged DataFrame
cols = ['inv_01_dc_current_inv_149579'	,'inv_01_dc_voltage_inv_149580',	'inv_01_ac_current_inv_149581',	'inv_01_ac_voltage_inv_149582', 'inv_01_ac_power_inv_149583', 'ambient_temperature_o_149575','meter_revenue_grade_ac_output_meter_149578','poa_irradiance_o_149574']
mergred_df_new = merged_df[cols]
```

```python
mergred_df_new.tail()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'merged_df' is the merged dataset
# Calculate the correlation matrix for numerical columns
corr_matrix = mergred_df_new.corr()

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Merged Dataset")
plt.show()
```
